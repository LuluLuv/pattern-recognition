```python
# Contabiliza etiquetas
def get_counters(col_name, label):
    full_list = ";".join(col_name)
    each_word = full_list.split(";")
    each_word = Counter(each_word).most_common()
    return pd.DataFrame(each_word, columns = [label, "Participantes"])

# Calcular porcentajes
def percent (row, col):
    count = df[col].count()
    return 100 * row[1] / count
```

Para casos como el presente, es conveniente visualizar los resultados del conjunto de variables en una sola gráfica. 

```python
def get_fig_bar_mult(columns, labels, x_data, angle_likert = 0):
    colors = random_colors(len(columns))

    traces = []
    for i in range(0, len(x_data[0])):
        for xd, yd in zip(x_data, labels):
            traces.append(go.Bar(
                x=[xd[i]],
                y=[yd],
                orientation='h',
                marker=dict(
                    color=colors[i],
                    line=dict(
                            color='rgb(248, 248, 249)',
                            width=1)
                )
            ))

    layout = go.Layout(
        xaxis=dict(
            showgrid=False,
            showline=False,
            showticklabels=False,
            zeroline=False,
            domain=[0.15, 1]
        ),
        yaxis=dict(
            showgrid=False,
            showline=False,
            showticklabels=False,
            zeroline=False,
        ),
        barmode='stack',
        paper_bgcolor='rgb(248, 248, 255)',
        plot_bgcolor='rgb(248, 248, 255)',
        margin=dict(
            l=120,
            r=10,
            t=140,
            b=80
        ),
        hovermode='closest',
        showlegend=False,
    )

    annotations = []
    for yd, xd in zip(labels, x_data):
        # labeling the y-axis
        annotations.append(dict(xref='paper', yref='y',
                                x=0.14, y=yd,
                                xanchor='right',
                                text=str(yd),
                                font=dict(family='Arial', size=14,
                                          color='rgb(67, 67, 67)'),
                                showarrow=False, align='right',
                               )
                          )
        # labeling the first Likert scale (on the top)
        if yd == labels[-1]:
            annotations.append(dict(xref='x', yref='paper',
                                    x=xd[0] / 2, y=1.1,
                                    text=top_labels[0],
                                    font=dict(family='Arial', size=14,
                                              color='rgb(67, 67, 67)'),
                                    showarrow=False,
                                    textangle = angle_likert
                                   )
                              )
        space = xd[0]
        for i in range(1, len(xd)):
                # labeling the Likert scale
                if yd == labels[-1]:
                    annotations.append(dict(xref='x', yref='paper',
                                            x=space + (xd[i]/2), y=1.1,
                                            text=top_labels[i],
                                            font=dict(family='Arial', size=14,
                                                      color='rgb(67, 67, 67)'),
                                            showarrow=False,
                                            textangle = angle_likert
                                           )
                                      )
                space += xd[i]
    layout['annotations'] = annotations

    return go.Figure(data=traces, layout=layout)
```

```python
import random

def random_colors(number_of_colors):
    color = ["#"+''.join([random.choice('0123456789ABCD') for j in range(6)])
                 for i in range(number_of_colors)]
    return color
```

```python
def get_fig_bar_group(data, labels, option = 'group'): #'group' / 'stack'
    traces = []
    colors = random_colors(len(data))
    for i in range(len(data)):
        traces.append(go.Bar(
            x=data[i].iloc[:,0],
            y=data[i].iloc[:,1],
            name=labels[i],
            marker=dict(
                color=colors[i]
            )
        ))
        
    layout = go.Layout(
        xaxis=dict(tickangle=15),
        barmode=option,
    )
    
    return go.Figure(data=traces, layout=layout)
```

```python
# Instalar el módulo squarify
import squarify

def get_fig_treemap(data):
    x = 0
    y = 0
    width = 50
    height = 50

    data = data.head(25)
    type_list = data.iloc[:,0]
    values = data.iloc[:,1]

    normed = squarify.normalize_sizes(values, width, height)
    rects = squarify.squarify(normed, x, y, width, height)
    
    colors = random_colors(20)
    shapes = []
    annotations = []
    counter = 0
    color = 0

    for r in rects:
        shapes.append( 
            dict(
                type = 'rect', 
                x0 = r['x'], 
                y0 = r['y'], 
                x1 = r['x']+r['dx'], 
                y1 = r['y']+r['dy'],
                line = dict( width = 1 ),
                fillcolor = colors[color]
            ) 
        )
        annotations.append(
            dict(
                x = r['x']+(r['dx']/2),
                y = r['y']+(r['dy']/2),
                text = "{}".format(type_list[counter]),
                showarrow = False
            )
        )
        counter = counter + 1
        color = color + 1
        if color >= len(colors):
            color = 0

        trace0 = go.Scatter(
            x = [ r['x']+(r['dx']/2) for r in rects ],
            y = [ r['y']+(r['dy']/2) for r in rects ],
            text = [ str(v) for v in values ], 
            mode = 'text',
        )

        layout = dict(
            height=600, 
            width=750,
            xaxis=dict(showgrid=False,zeroline=False),
            yaxis=dict(showgrid=False,zeroline=False),
            shapes=shapes,
            annotations=annotations,
            hovermode='closest',
            font=dict(color="#FFFFFF"),
            margin = go.layout.Margin(
                l=0,
                r=0,
                pad=0
            )
        )

    return dict(data=[trace0], layout=layout)

figure = get_fig_treemap(data)
py.offline.iplot(figure)
```

```python
def get_counters_simple(col, label):
    full_list = "|".join(col)
    each_word = full_list.split("|")
    each_word = Counter(each_word).most_common()
    return pd.DataFrame(each_word, columns = [label, "Participantes"])

print(df['Gender'].describe())
data = get_counters_simple(df['Gender'].dropna(), "Género")
display(HTML(data.to_html()))
```

El problema evidente, en este conjunto de datos, es el de valores faltantes. Para determinar la magnitud del problema, calculamos el porcentaje de valores faltantes en cada columna: Un porcentaje superior al 5% será un problema serio.

```python
missing = {}
for col in df.describe(include='all'):
    missing[col] = 100*sum(pd.isnull(df[col])) / df.shape[0]
```

```python
from operator import itemgetter
table = sorted(missing.items(), key=itemgetter(1))

data = pd.DataFrame(table, columns=["Columna", "Valores Faltantes (%)"])
display(HTML(data.to_html()))
```
